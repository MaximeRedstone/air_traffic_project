```python
import pandas as pd
from merge_transformer import MergeTransformer
```

```python
waiting_time = pd.read_csv('../../data/WAITING_TIME.csv', sep=';')
```

```python
waiting_time['Date'] = pd.to_datetime(waiting_time['Date'])
```

```python
total_passengers = waiting_time.groupby(["Date", 'Airport'])["Total"].sum()
total_booths = waiting_time.groupby(["Date", 'Airport'])["Flights"].sum()
total_flights = waiting_time.groupby(["Date", 'Airport'])["Booths"].sum()
```

```python
total_passengers = pd.DataFrame(total_passengers).reset_index(drop=False)
total_booths = pd.DataFrame(total_booths).reset_index(drop=False)
total_flights = pd.DataFrame(total_flights).reset_index(drop=False)
```

```python
merge_transform = MergeTransformer(
    X_ext=total_booths,
    how='left', on=['Date', 'Airport'])

total_passengers = merge_transform.fit_transform(total_passengers)

merge_transform = MergeTransformer(
    X_ext=total_flights,
    how='left', on=['Date', 'Airport'])

total_passengers = merge_transform.fit_transform(total_passengers)
```

```python
total_passengers
```

```python
total_passengers['Mean per flight'] = total_passengers['Total'] / total_passengers['Flights']
```
